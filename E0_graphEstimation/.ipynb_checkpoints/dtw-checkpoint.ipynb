## Class-specific graphs

```python
for c in range(len(folders)):

    ########################
    ###### PARAMETROS ######
    ########################
    numberOfTimeStep = 14
    norm = "normPower2"
    ########################

    ####### BEGIN CODE ===>>
    # Load data
    X_train = np.load("../DATA/" + folders[c] + "/X_train_tensor_"+norm+".npy")
    y_train = pd.read_csv("../DATA/" + folders[c] + "/y_train_"+norm+".csv")
    
    #####################################
    number_of_pat = []
    for i in range(numberOfTimeStep):
        df = pd.DataFrame(X_train[:,i])

        filtered_df = df[~(df == 666).any(axis=1)]

        number_of_pat.append(filtered_df.shape[0])

    plt.figure(figsize = (15, 8))
    plt.bar(np.arange(1, numberOfTimeStep+1, 1), number_of_pat)
    plt.xticks(fontsize=12, rotation=90)
    plt.yticks(fontsize=12)
    plt.grid(axis='y',linestyle='--', alpha=0.7)
    plt.title("Numero de pacientes por instantes temporales")
    plt.show()
    #####################################

    # Reshape y_train
    y_train_aux = y_train.iloc[0:y_train.shape[0]:numberOfTimeStep].reset_index(drop=True)
    
    # Get data by label (AMR and non-AMR patients)
    AMR_idx = list(y_train_aux[y_train_aux.individualMRGerm_stac == 1].index)
    X_AMR_tr = X_train[AMR_idx]
        
    noAMR_idx = list(y_train_aux[y_train_aux.individualMRGerm_stac == 0].index)
    X_noAMR_tr = X_train[noAMR_idx]
    
    print("X_noAMR_tr shape before reshape and val:", X_noAMR_tr.shape)
    print("X_AMR_tr shape before reshape and val:", X_AMR_tr.shape)
    
    # Reshape data by features --> PxTxF to FxTxP
    X_AMR_tr = utils.reshape_patients_by_features(X_AMR_tr, keys, numberOfTimeStep)
    X_noAMR_tr = utils.reshape_patients_by_features(X_noAMR_tr, keys, numberOfTimeStep)
    
    print("X_AMR_tr shape after reshape:", X_AMR_tr.shape)
    print("X_noAMR_tr shape after reshape:", X_noAMR_tr.shape)
    
    result = Parallel(n_jobs=30)(delayed(dtw_hgd.compute_hgd_matrix)(X_AMR_tr, X_AMR_tr[f1,:,:], X_AMR_tr[f2,:,:], keys[f1], keys[f2], binary, continues)
             for f1 in range(X_AMR_tr.shape[0]) for f2 in range(X_AMR_tr.shape[0]))
    
    lista_sin_nan = dtw_hgd.replace_nan_with_zero(result)
    dtw_matrix_train = Parallel(n_jobs=30)(
            delayed(dtw_hgd.dtw)(lista_sin_nan[i])
            for i in range(X_AMR_tr.shape[0]*X_AMR_tr.shape[0])
        )

    data_amr = np.array(dtw_matrix_train).reshape((X_AMR_tr.shape[0], X_AMR_tr.shape[0]))


    pd.DataFrame(data_amr,columns=keys, index=keys).to_csv("../E1_InferenceOverGraph/dtw_matrices/"+folders[c]+"/tr_AMR_"+norm+".csv", index=False)
        
    
    result = Parallel(n_jobs=30)(delayed(dtw_hgd.compute_hgd_matrix)(X_noAMR_tr, X_noAMR_tr[f1,:,:], X_noAMR_tr[f2,:,:], keys[f1], keys[f2], binary, continues)
             for f1 in range(X_noAMR_tr.shape[0]) for f2 in range(X_noAMR_tr.shape[0]))

    
    lista_sin_nan = dtw_hgd.replace_nan_with_zero(result)
    dtw_matrix_train = Parallel(n_jobs=30)(
            delayed(dtw_hgd.dtw)(lista_sin_nan[i])
            for i in range(X_noAMR_tr.shape[0]*X_noAMR_tr.shape[0])
        )

    data_noamr = np.array(dtw_matrix_train).reshape((X_noAMR_tr.shape[0], X_noAMR_tr.shape[0]))
    
    pd.DataFrame(data_noamr,columns=keys, index=keys).to_csv("../E1_InferenceOverGraph/dtw_matrices/"+folders[c]+"/tr_noAMR_"+norm+".csv", index=False)
```

## Single adjacency matrix capturing the whole population

```python
for c in range(len(folders)):

    ########################
    ###### PARAMETROS ######
    ########################
    numberOfTimeStep = 14
    norm = "normPower2"
    ########################

    ####### BEGIN CODE ===>>
    # Load data
    X_train = np.load("../DATA/" + folders[c] + "/X_train_tensor_"+norm+".npy")
    
    #####################################
    number_of_pat = []
    for i in range(numberOfTimeStep):
        df = pd.DataFrame(X_train[:,i])

        filtered_df = df[~(df == 666).any(axis=1)]

        number_of_pat.append(filtered_df.shape[0])

    plt.figure(figsize = (15, 8))
    plt.bar(np.arange(1, numberOfTimeStep+1, 1), number_of_pat)
    plt.xticks(fontsize=12, rotation=90)
    plt.yticks(fontsize=12)
    plt.grid(axis='y',linestyle='--', alpha=0.7)
    plt.title("Numero de pacientes por instantes temporales")
    plt.show()
    #####################################

    
    print("X_train shape before reshape and val:", X_train.shape)
    
    # Reshape data by features --> PxTxF to FxTxP
    X_train = utils.reshape_patients_by_features(X_train, keys, numberOfTimeStep)
    
    print("X_train shape after reshape:", X_train.shape)
    
    result = Parallel(n_jobs=-1)(delayed(dtw_hgd.compute_hgd_matrix)(X_train, X_train[f1,:,:], X_train[f2,:,:], keys[f1], keys[f2], binary, continues)
             for f1 in range(X_train.shape[0]) for f2 in range(X_train.shape[0]))
    
    lista_sin_nan = dtw_hgd.replace_nan_with_zero(result)
    dtw_matrix_train = Parallel(n_jobs=30)(
            delayed(dtw_hgd.dtw)(lista_sin_nan[i])
            for i in range(X_train.shape[0]*X_train.shape[0])
        )

    data_amr = np.array(dtw_matrix_train).reshape((X_train.shape[0], X_train.shape[0]))


    pd.DataFrame(data_amr, columns=keys, index=keys).to_csv("../E2_Standard-GCN/dtw_matrices/"+folders[c]+"/tr_AMR_"+norm+".csv", index=False)

    pd.DataFrame(data_amr,columns=keys, index=keys).to_csv("../E3_SpatioTemporal-GCN/dtw_matrices/"+folders[c]+"/tr_AMR_"+norm+".csv", index=False)
    
```
